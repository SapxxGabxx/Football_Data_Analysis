This chances are the expected goals.

Now we have to filter by ID. So we choice the Russia-Denmark json file with <b>ID=3788769<b>.

```python
with open('C:/Users/sapga/OneDrive/Desktop/open-data-master/data/events/3788769.json', 'r', encoding='utf-8') as f:
    rus_den = json.load(f)
    
rus_den
```

Now we use <b>json_normalize<b> function from pandas library. It is so powerful because it can handle the nested JSON.

```python
df = pd.json_normalize(rus_den, sep='_').assign(match_id="3788769")
pd.set_option('display.max_columns',200)
df.head()
```

Now we want to get the shots. So we filter by type_name='Shot'.

```python
shots = df[df.type_name == 'Shot'].set_index('id')
pd.set_option('display.max_columns',200)
shots.head(5)
```

Now we use <b>FCPython</b> from the <b>createPitch</b> library to draw the pitch. On this map, we want to see how many shots each team has made. Also, we want to know how huge the chance for scoring a goal is. We call this chance the expected goals.

```python
import numpy as np
import matplotlib.pyplot as plt
from FCPython import createPitch

pitch_width = 120
pitch_height = 80

fig, ax = createPitch(pitch_width, pitch_height, 'yards', 'gray')

home_team = 'Russia'
away_team = 'Denmark'

for i, shot in shots.iterrows():
    x = shot['location'][0]
    y = shot['location'][1]
    
    goal = shot['shot_outcome_name']=='Goal'
    team_name = shot['team_name']
    
    circle_size = 2
    circle_size = np.sqrt(shot['shot_statsbomb_xg'] * 15)
    
    if team_name == home_team:
        if goal:
            shot_circle = plt.Circle((x, pitch_height-y), circle_size, color='red')
            plt.text((x+1), pitch_height-y+1, shot['player_name'])
        else:
            shot_circle = plt.Circle((x, pitch_height-y), circle_size, color='red')
            shot_circle.set_alpha(.2)

    if team_name == away_team:
        if goal:
            shot_circle = plt.Circle((pitch_width-x, y), circle_size, color='blue')
            plt.text((pitch_width-x+1), y+1, shot['player_name'])
        else:
            shot_circle = plt.Circle((pitch_width-x, y), circle_size, color='blue')
            shot_circle.set_alpha(.2)
    
    ax.add_patch(shot_circle)
    
plt.text(5, 75, away_team + ' shots')
plt.text(80, 75, home_team + ' shots')

plt.title('Russia vs Denmark at 2021 Uefa Euro Cup')

fig.set_size_inches(20, 14)
fig.savefig('rus_den_shots.png', dpi=300) 

plt.show()
```

I created a collection of dots that correspond to the shots that have been made.

For generating the dots, we set the loop for iterating rows inside the dataframe. 

Circles are generated on the basis of coordinates and xG.
The size of the circle depends on the value of xG. Furthermore, if the shot is a goal it will be of the established colour, otherwise if it is not a goal the colour will be opaque.

As we can see from the picture, Denmark deservedly won the match with 4 goals, as they created many more chances, many of them dangerous, i.e. shots with a high xG. In contrast Russia created few chances scoring only one goal.
